```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    vif=pd.DataFrame()
    vif["variables"]=X.columns
    vif["VIF"]=[variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
    return (vif)
```

```python
calc_vif(dataset[quan])
```

```python
calc_vif(dataset[["Monthly_Sales", "Rent"]])
```

## T-Test & Hypothesis test

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
city4=dataset[dataset['City'] == 4]["Footfall_per_day"]
city3=dataset[dataset['City'] == 3]["Footfall_per_day"]
ttest_ind(city4,city3)
```

```python
#Statement: Test whether Footfall_per_day of city 3 and city 4 are the same.
#Null Hypothesis (H₀): There is no significant difference in Footfall per day between City 4 and City 3.
#Accept null hypothesis reject alternate hypothesis.
```

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
city4=dataset[dataset['City'] == 4]["Nearby_Brands"]
city3=dataset[dataset['City'] == 3]["Nearby_Brands"]
ttest_ind(city4,city3)
```

```python
# Nearby brands are almost significantly different, but still not statistically significant at the 5% level (p ≈ 0.052 > 0.05).
```

```python
from scipy.stats import ttest_ind
dataset=dataset.dropna()
Mall_Name3=dataset[dataset['Mall_Name']==3]["Monthly_Sales"]
Mall_Name4=dataset[dataset['Mall_Name']==4]["Monthly_Sales"]
ttest_ind(Mall_Name3,Mall_Name4)
```

```python
#Statement: Test whether Monthly Sales of Mall 3 and Mall 4 are the same.
#Null Hypothesis (H₀): There is no significant difference in Monthly Sales between Mall 3 and Mall 4.
#Alternative Hypothesis (H₁): There is a significant difference in Monthly Sales between Mall 3 and Mall 4.

#Reject null Hypothesis p<0.05

#Accept null hypothesis reject alternate hypothesis.
```
