```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("preprocessed_restaurant_data.csv")
dataset
```

```python
dataset.shape
```

```python
sns.distplot(dataset["Revenue"],hist=True,kde=True,kde_kws={'color':'purple'})
```

```python
sns.jointplot(x = 'Social Media Followers', y = 'Revenue', data = dataset, kind ='hex', color='green')
plt.show()
```

```python
sns.pairplot(dataset[['Revenue', 'Rating', 'Average Meal Price', 'Service Quality Score']])
plt.show()
```

```python
sns.boxplot(x=dataset['Revenue'])
```

```python
sns.boxplot(x=dataset['Marketing Budget'])
```

```python
numeric_cols = ['Rating', 'Seating Capacity', 'Average Meal Price',
                'Marketing Budget', 'Social Media Followers',
                'Chef Experience Years', 'Number of Reviews',
                'Ambience Score', 'Service Quality Score', 'Revenue']

dataset[numeric_cols].hist(figsize=(15,12), bins=30)
plt.tight_layout()
plt.show()
```

```python
sns.histplot(dataset[dataset['Location_Rural']==1]['Revenue'], bins=30, kde=True)
plt.title('Revenue Distribution - Rural')
plt.show()
```

```python
sns.histplot(dataset[dataset['Location_Suburban']==1]['Revenue'], bins=30, kde=True)
plt.title('Revenue Distribution - Suburban')
plt.show()
```
