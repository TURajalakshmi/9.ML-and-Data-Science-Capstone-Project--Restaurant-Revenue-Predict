```python
from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(X):
    vif=pd.DataFrame()
    vif["variables"]=X.columns
    vif["VIF"]=[variance_inflation_factor(X.values, i) for i in range (X.shape[1])]
    return (vif)
```

```python
calc_vif(Dataset[quan])
```

## T-Test & Hypothesis test

```python
from scipy.stats import ttest_ind
Dataset = Dataset.dropna()
Indian = Dataset[Dataset['Cuisine_Indian'] == 1]["Revenue"]
Mexican = Dataset[Dataset['Cuisine_Mexican'] == 1]["Revenue"]
ttest_ind(Indian,Mexican)
```

```python
# Statement: Test whether the Revenue of Cuisine_Indian and Cuisine_Mexican are the same.
# Test Result: p-value = 4.04499 Ã— 10â»â´â° (p < 0.05) Statistic = 13.47
# Since the p-value is far below 0.05, Reject the null hypothesis (Hâ‚€).  Accept the alternative hypothesis (Hâ‚)
# There is a statistically significant difference in Revenue between Indian and Mexican cuisine restaurants.
```

```python
from scipy.stats import ttest_ind
Dataset = Dataset.dropna()
Rural = Dataset[Dataset['Location_Rural'] == 1]["Marketing Budget"]
Suburban = Dataset[Dataset['Location_Suburban'] == 1]["Marketing Budget"]
ttest_ind(Rural,Suburban)
```

```python
# Nearby restaurant locations (Rural vs Suburban) show no statistically significant difference in marketing budget.p-value â‰ˆ 0.369 (p > 0.05).
# Accept the null hypothesis (Hâ‚€).  Reject the alternate hypothesis (Hâ‚).
```

```python
from scipy.stats import ttest_ind
Dataset=Dataset.dropna()
Weekend = Dataset[Dataset['Weekend Reservations'] > Dataset['Weekday Reservations']]['Average Meal Price']
Weekday = Dataset[Dataset['Weekday Reservations'] > Dataset['Weekend Reservations']]['Average Meal Price']
ttest_ind(Weekend, Weekday, equal_var=False)
```

```python
# p-value = 0.3519 â†’ > 0.05. Accept Hâ‚€. Reject H.
# No significant difference (Weekend vs Weekday â†’ No significant difference).
```

```python
from scipy.stats import ttest_ind
Dataset=Dataset.dropna()
Parking_Yes = Dataset[Dataset['Parking Availability_Yes'] == 1]["Revenue"]
Parking_No = Dataset[Dataset['Parking Availability_Yes'] == 0]["Revenue"]
ttest_ind(Parking_Yes,Parking_No)
```

```python
# Statement: Test whether the average revenue of restaurants with parking and restaurants without parking are the same.
# Null Hypothesis (Hâ‚€): There is no significant difference in average revenue between restaurants with Parking and restaurants without Parking.
# Alternate Hypothesis (Hâ‚): There is a significant difference in average revenue between restaurants with Parking and restaurants without Parking.
# p-value = 0.7456 (p > 0.05)
# Accept the null hypothesis (Hâ‚€). Reject the alternate hypothesis (Hâ‚).
```
