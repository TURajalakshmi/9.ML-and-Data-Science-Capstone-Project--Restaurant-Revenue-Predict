```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("preprocessed_mall_data.csv")
dataset
```

```python
dataset.shape
```

```python
sb.distplot(dataset["Rent"],hist=True,kde=True,kde_kws={'color':'purple'})
```

```python
sb.jointplot(x = 'Monthly_Sales', y = 'Rent', data = dataset)
plt.show()
```

```python
sb.jointplot(x = 'Footfall_per_day', y = 'Rent', data = dataset, kind ='hex', color='green')
plt.show()
```

```python
sb.jointplot(x = 'Shop_Size_sqft', y = 'Rent', data = dataset, kind='kde', color='purple')
plt.show()
```

```python
sb.pairplot(dataset[['Floor','Shop_Size_sqft','Footfall_per_day','Nearby_Brands',
             'Has_Food_Court','Monthly_Sales','Rent']],hue='Rent', diag_kind='kde', kind='scatter', palette = 'husl')
```
